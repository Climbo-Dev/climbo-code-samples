## Activation Functions

```python
%matplotlib inline
from itertools import product
import torch
from torch import nn
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['figure.dpi'] = 200

activations = [nn.Tanh, nn.Sigmoid, nn.ReLU, nn.SiLU, nn.GELU, nn.ReLU6, nn.Hardtanh, nn.Softplus, nn.Softsign]

# Configuration
xmin = -8.
xmax = 8.
ncols = 3
x = torch.arange(xmin, xmax, 0.5)

# Render Plots
nacts = len(activations)
nrows = nacts // ncols + 1

figure, axis = plt.subplots(nrows, ncols, layout="constrained")

for i, (irow, icol) in enumerate(product(range(nrows), range(ncols))):
    ax = axis[irow, icol]
    if i >= nacts:
        ax.set_visible(False)
    else:
        Func = activations[i]
        f = Func()
        ax.axhline(color='r', ls='--')
        ax.axvline(color='r', ls='--')
        ax.plot(x, f(x))
        ax.set_title(Func.__name__)
```
